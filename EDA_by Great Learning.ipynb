# Diabetes Detection using MVA and Machine Learning

```python
# Importing basic packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_profiling import ProfileReport
from subprocess import check_output
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn import metrics
%matplotlib inline
from sklearn.metrics import classification_report, accuracy_score
```

### Import the dataset

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.head()
```

### Details about the dataset

```python
df.info()
```

### statistical aspect of the dataset

```python
df.describe()
```

### Preprocessing  - Checking for NaN values

```python
df.isna().any()
```

### Understanding correlation between the labels

```python
# Correlation
correlation_metrics = df.corr()
fig = plt.figure(figsize=(14,9))
sns.heatmap(correlation_metrics, vmax = .9, square=True) #annot = value
plt.show()
print(correlation_metrics)
```

### number of people with no diabetes

```python
No_Sugar = 0
result = df['Outcome']
for i in range(len(result)):
    if result[i] == 0:
        No_Sugar = No_Sugar+1
print(No_Sugar)
        
```

### number of people with diabetes

```python
Sugar = 0
result = df['Outcome']
for i in range(len(result)):
    if result[i] == 1:
        Sugar = Sugar+1
print(Sugar)
```

### Printing count plot with Seaborn

```python
# creating count plot with title using seaborn
sns.countplot(x=df.Outcome)
plt.title("No diabetes Vs Diabetes")
```

### Splitting the training and test dataset

```python
x=df.drop(['Outcome'],axis=1) # Drop the target variable
y=df['Outcome']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

### Using Logistic Regression

```python
L=LogisticRegression()
L.fit(x_train,y_train)
```

### Performing predictions

```python
y_pred = L.predict(x_test)
```

### Confusion Matrix

```python
confusion_matrix = metrics.confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

### Final accuracy score

```python
accuracy = L.score(x_test,y_test)
print('Accuracy score of the Logistic regression model: ', accuracy*100, '%')
```

      Removing these missing data from the dataset will result in information loss and it is not

### Outlier Detection and treatment

#### Outlier Detection
     Outliers are extreme values that deviate from other observations on data, they may indicate
     a variability in a measurement, experimental error or a novelty.
     Boxplots are a great way of detecting outliers.Once the outliers have been detected, they can
     be imputed with the 5th and 95th percentiles.

```python
# Outlier detection using boxplots.
plt.figure(figsize=(20,15))
plt.subplot(4,4,1)
sns.boxplot(df['Pregnancies'])

plt.subplot(4,4,2)
sns.boxplot(df['Glucose'])

plt.subplot(4,4,3)
sns.boxplot(df['BloodPressure'])

plt.subplot(4,4,4)
sns.boxplot(df['SkinThickness'])

plt.subplot(4,4,5)
sns.boxplot(df['Insulin'])

plt.subplot(4,4,6)
sns.boxplot(df['BMI'])

plt.subplot(4,4,7)
sns.boxplot(df['DiabetesPedigreeFunction'])

plt.subplot(4,4,8)
sns.boxplot(df['Age'])
```

##### Apart from 'Glucose' all the other attributes show presence of outliers. These lower level and upper level outliers will be replaced by the 5th and 95th percentile respectively.  

### IQR
The interquartile range(IQR) is a measure of variability, based on dividing datset into quartiles.
Quartiles divide a rank-ordered datset into four equal parts. The values that divide each part are called the first, second, and third quartiles; and they are denoted by Q1, Q2 and Q3 respectively.

##### Numpy.clip() function is used to Clip(limit) the values in an array.
##### Syntax : numpy.clip(a, a_min, a_max, out=None)
##### Parameters:
##### a: Array containing elements to clip.
##### a_min: Minimum value
##### -> If None, clipping is not performed on lower interval edge. Not more than one of a_min and a_max be None
##### a_max: Maximum value

```python
df['Pregnancies']=df['Pregnancies'].clip(lower=df['Pregnancies'].quantile(0.05), upper=df['Pregnancies'].quantile(0.95))
df['BloodPressure']=df['BloodPressure'].clip(lower=df['BloodPressure'].quantile(0.05), upper=df['BloodPressure'].quantile(0.95))
df['SkinThickness']=df['SkinThickness'].clip(lower=df['SkinThickness'].quantile(0.05), upper=df['SkinThickness'].quantile(0.95))
df['Insulin']=df['Insulin'].clip(lower=df['Insulin'].quantile(0.05), upper=df['Insulin'].quantile(0.95))
df['BMI']=df['BMI'].clip(lower=df['BMI'].quantile(0.05), upper=df['BMI'].quantile(0.95))
df['DiabetesPedigreeFunction']=df['DiabetesPedigreeFunction'].clip(lower=df['DiabetesPedigreeFunction'].quantile(0.05), upper=df['DiabetesPedigreeFunction'].quantile(0.95))
df['Age']=df['Age'].clip(lower=df['Age'].quantile(0.05), upper=df['Age'].quantile(0.95))
```

```python
# Lest visualise the boxplots after imputing the outliers.
plt.figure(figsize=(20,15))
plt.subplot(4,4,1)
sns.boxplot(df['Pregnancies'])

plt.subplot(4,4,2)
sns.boxplot(df['Glucose'])

plt.subplot(4,4,3)
sns.boxplot(df['BloodPressure'])

plt.subplot(4,4,4)
sns.boxplot(df['SkinThickness'])

plt.subplot(4,4,5)
sns.boxplot(df['Insulin'])

plt.subplot(4,4,6)
sns.boxplot(df['BMI'])

plt.subplot(4,4,7)
sns.boxplot(df['DiabetesPedigreeFunction'])

plt.subplot(4,4,8)
sns.boxplot(df['Age'])
```

```python
# As we can see, there are still outliers in columns SkinThickness and Insulin. Lets try manipulating the percentile values.
df['SkinThickness']=df['SkinThickness'].clip(lower=df['SkinThickness'].quantile(0.03), upper=df['SkinThickness'].quantile(0.97))
df['Insulin']=df['Insulin'].clip(lower=df['Insulin'].quantile(0.21), upper=df['Insulin'].quantile(0.79))
```

```python
plt.figure(figsize=(20,15))

plt.subplot(4,4,1)
sns.boxplot(df['SkinThickness'])

plt.subplot(4,4,2)
sns.boxplot(df['Insulin'])
```

```python
# As we can see, there are still outliers in columns SkinThickness and Insulin. Lets try manipulating the percentile values.
df['SkinThickness']=df['SkinThickness'].clip(lower=df['SkinThickness'].quantile(0.1), upper=df['SkinThickness'].quantile(0.9))
df['Insulin']=df['Insulin'].clip(lower=df['Insulin'].quantile(0.25), upper=df['Insulin'].quantile(0.75))
```

```python
plt.figure(figsize=(20,15))

plt.subplot(4,4,1)
sns.boxplot(df['SkinThickness'])

plt.subplot(4,4,2)
sns.boxplot(df['Insulin'])
```
